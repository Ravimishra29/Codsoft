```python
from tkinter import *

t = Tk()
t.title("Calculator")
t.geometry("580x610")
t.resizable(False,False)
t.configure(bg="black")

equation = ""

def show(value):
    global equation
    equation+=value
    r.config(text=equation)

def clear():
    global equation
    equation = ""
    r.config(text=equation)

def calculate():
    global equation
    result = ""
    if equation !="":
        try:
            result = eval(equation)
        except:
            result = "error"
            equation = ""
    r.config(text=result)

r = Label(t,width=25,height=2,text="",fg="white",bg="black",font=("arial",29))
r.pack()

Button(t,text="AC", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Red",command=lambda: clear()).place(x=430,y=500)
Button(t,text="/", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="black", bg="white",command=lambda: show("/")).place(x=150,y=100)
Button(t,text="%", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="black", bg="white",command=lambda: show("%")).place(x=290,y=100)
Button(t,text="*", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="black", bg="white",command=lambda: show("*")).place(x=430,y=100)

Button(t,text="7", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Gray",command=lambda: show("7")).place(x=10,y=200)
Button(t,text="8", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Gray",command=lambda: show("8")).place(x=150,y=200)
Button(t,text="9", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Gray",command=lambda: show("9")).place(x=290,y=200)
Button(t,text="-", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="black", bg="white",command=lambda: show("-")).place(x=430,y=200)

Button(t,text="4", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Gray",command=lambda: show("4")).place(x=10,y=300)
Button(t,text="5", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Gray",command=lambda: show("5")).place(x=150,y=300)
Button(t,text="6", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Gray",command=lambda: show("6")).place(x=290,y=300)
Button(t,text="+", width=5, height=3, font=("arial",31,"bold"), bd=5,fg="black", bg="white",command=lambda: show("+")).place(x=430,y=300)

Button(t,text="1", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Gray",command=lambda: show("1")).place(x=10,y=400)
Button(t,text="2", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Gray",command=lambda: show("2")).place(x=150,y=400)
Button(t,text="3", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Gray",command=lambda: show("3")).place(x=290,y=400)
Button(t,text="0", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Gray",command=lambda: show("0")).place(x=150,y=500)

Button(t,text="00", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="black", bg="white",command=lambda: show("00")).place(x=10,y=100)
Button(t,text=".", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="Gray",command=lambda: show(".")).place(x=10,y=500)
Button(t,text="=", width=5, height=1, font=("arial",31,"bold"), bd=5,fg="white", bg="blue",command=lambda: calculate()).place(x=290,y=500)

t.mainloop()
```
