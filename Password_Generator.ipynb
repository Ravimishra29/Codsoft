```python
from tkinter import *
import random, string

t = Tk()
t.geometry("500x450")
t.configure(bg="sky blue")
t.title("Password Generator")
t.resizable(False,False)


title = StringVar()
title.set("Strength level of Password:")
label = Label(t, textvariable=title,font=("arial",20,"bold"),bd=5).pack()

def selection():
    selection = choice.get()

choice = IntVar()
R1 = Radiobutton(t, text="BASIC LEVEL",variable=choice,value=1,font=("arial",10,"bold"),bd=10,bg="Pink",command=selection).pack(anchor=CENTER)
R2 = Radiobutton(t, text="MEDIUM LEVEL",variable=choice,value=2,font=("arial",10,"bold"),bd=10,bg="Pink",command=selection).pack(anchor=CENTER)
R3 = Radiobutton(t, text="ADVANCED LEVEL",variable=choice,value=3,font=("arial",10,"bold"),bd=10,bg="Pink",command=selection).pack(anchor=CENTER)

labelchoice=Label(t)
labelchoice.pack()

label = StringVar()
label.set("Password length:")
lentitle = Label(t, textvariable=label,font=("arial",20,"bold"),bd=5).pack()

val = IntVar()
spinlenght = Spinbox(t, from_=6, to_=20,font=("arial",15,"bold"),bd=5, textvariable=val, width=18).pack()


def callback():
    lsum.config(text=passgen())

genButton=Button(t,text="Generate password", font=("arial",10,"bold"),bd=10, height=2,fg="white",bg="black", command=callback, pady=10)
genButton.pack()
Password=str(callback)

lsum=Label(t,text="",width=20,font=("arial",20,"bold"))
lsum.pack(side=BOTTOM)


poor = string.ascii_uppercase + string.ascii_lowercase
average = string.ascii_uppercase + string.ascii_lowercase + string.digits
symbols = """~!@#$%^&*()_+=<>'''?.;:"""
advance = poor + average + symbols

def passgen():
    if choice.get() == 1:
        return"".join(random.sample(poor, val.get()))
    elif choice.get() == 2:
        return"".join(random.sample(average, val.get()))
    elif choice.get() == 3:
        return"".join(random.sample(advance, val.get()))
    
t.mainloop()
```
